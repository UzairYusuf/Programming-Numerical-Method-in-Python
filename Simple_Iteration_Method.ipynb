```python
x = 0 #the intial guess
for iteration in range(1,101): #100 iteration
  xnew = (2*x**2 + 3)/5
  print(iteration, x)
  if abs(xnew - x) < 0.000001:
    break
  x = xnew
print('The root : %0.5f' % xnew)
print('The number of iteration : %d' % iteration)
```

```python
from math import sqrt
x = 1.6 #the intial guess
for iteration in range(1,101): #100 iteration
  xnew = sqrt((5*x-3)/2)
  print(x)
  if abs(xnew - x) < 0.000001:
    break
  x = xnew
print('The root : %0.5f' % xnew)
print('The number of iteration : %d' % iteration)
```
